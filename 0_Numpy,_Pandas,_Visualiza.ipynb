```python
#Dataframe
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/ven-27/datasets/master/titanic.csv")
df.head()

# Creo la tabla a graficar usando GROUPBY()
df_x=df[['Survived','Pclass']].groupby(by=['Pclass']).count().reset_index()
df_x['Pclass']=df_x['Pclass'].astype('str')
df_x.head()
```

```python
# Importo Libreria
import bokeh
from bokeh.palettes import Spectral3
from bokeh.io import show, output_notebook
from bokeh.models import CategoricalColorMapper, ColumnDataSource, FactorRange
from bokeh.plotting import figure
from bokeh.models import HoverTool, ColumnarDataSource
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import output_file

# Genero Grafico
output_notebook()
source = ColumnDataSource(data=dict(Pclass=df_x.Pclass, counts=df_x.Survived, color=Spectral3))
p = figure(x_range=df_x.Pclass, height=250, title="Clases involucradas")
p.vbar(x='Pclass', top='counts', width=0.9, color='color', legend_label="Pclass", source=source)

#Personalizacion
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Clase"
p.yaxis.axis_label = "Frecuencia"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"
p.add_tools(HoverTool())

# Guardar el grafico
# output_file("stacked.html")

#Mostrar grafico
show(p)
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Creo la tabla a graficar usando GROUPBY()
tips_bokeh=tips[['tip','day']].groupby(by=['day']).count().reset_index()

# Transformo en string la categoria a ubicar en el eje x
tips_bokeh['day']=tips_bokeh['day'].astype('str')

tips_bokeh.head()
```

```python
# Importo Libreria
import bokeh # Importo Libreria
from bokeh.palettes import Spectral8 #Paleta de colores (se puede cambiar el 6 por otro numero)
from bokeh.io import show, output_notebook
from bokeh.models import CategoricalColorMapper, ColumnDataSource, FactorRange
from bokeh.plotting import figure
from bokeh.models import HoverTool, ColumnarDataSource #Para Interctuar
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import output_file

#df= 'tips_bokeh'
#variable categorica eje x = 'day'
#variable numerica eje x = 'tip'

# Genero Grafico
source = ColumnDataSource(data=dict(day=tips_bokeh.day, counts=tips_bokeh.tip, color=Spectral8))
p = figure(x_range=tips_bokeh.day, height=400, title="PROPINA POR DIA DE LA SEMANA")
p.vbar(x='day', top='counts', width=0.7, color='color', legend_label="Propina", source=source)
output_notebook() #NECESARIO!

#Personalizacion
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "DIA DE LA SEMANA"
p.yaxis.axis_label = "PROPINA / TIPS"
p.legend.orientation = "horizontal"
p.legend.location = "top_left"
p.add_tools(HoverTool()) #Interactuar

# Guardar el grafico
output_file("stacked.html")

#Mostrar grafico
show(p)
```
